````python
import statistics

class Turing():
  def __init__(self, dados):
    self.dados = dados

  def max():
    return max(self.dados)

  def min(self):
    return min(self.dados)

  def media(self):
    return statistics.mean(self.dados)

  def quartis(self):
    return statistics.quantiles(self.dados)


````
