### 3. Utilizando apenas Pandas e Numpy, aplique a transformação de MinMaxScaling para cada uma das colunas numéricas do dataset.

Min-Max Scaling:

$$  x_{new} = \frac{x - x_{min}}{x_{max} - x_{min}}  $$

```python
colunas_numericas = df.select_dtypes(include=['int', 'float'])
for col in colunas_numericas:
  val_max = df[col].max()
  val_min = df[col].min()
  df[col] = (df[col]- val_min)/(val_max - val_min)
```

### 4. One Hot Encoding From Scratch

Quando trienamos um modelo de Machine Learning, só trabalhamos com valores numéricos, de forma que dados categóricos devem ser convertidos para números para que possa ser intempretados. Dentre as estratégias que existem para realizar essa tarefa, uma das mais populares é o **One-Hot-Encoding**. Nesse processo, as colunas categóricas são divididas em diversas features com valores de 0 e 1 (o número de features depende da quantidade de valores únicos na coluna original), como mostrado abaixo:

```python
departamento_df = pd.get_dummies(df['Department'], columns = ['Department'], dtype = int)
```

```python
departamento_df.head()

# Repare que, para um dado funcionário, a coluna correspondente a seu departamento de trabalho recebe o valor 1,
# enquanto os demais departamentos recebem o valor 0.
```
